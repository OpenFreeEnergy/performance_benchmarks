```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Normal: MD': 2578.7275586128235,
    'Hybrid: MD': 2733.7484908103943,
    'Hybrid: OMMTools HREX': 3806.9737679958344,
    'Hybrid: Manual HREX': 2781.1914410591125,
}
```

```python
df = pd.DataFrame.from_dict(data, orient='index', columns=['Time'])
df
```

```python
ax = df.plot(kind='bar')
ax.set_xticklabels(data.keys(), rotation=360)
ax.get_legend().remove()
ax.tick_params(axis='x', labelrotation=15)
plt.ylabel('Total time (seconds)')
plt.yticks([0, 1000, 2000, 3000, 4000])
plt.title('Time to complete 2.4M steps (MD = 1 replica, HREX = 12 replica)')
plt.savefig('sampler_analysis.png', bbox_inches='tight')
```
