```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Normal: MD': 2616.39,
    'Hybrid: MD': 2738.43,
    'Hybrid: HREX': 3896.63,
    'Hybrid: HREX-Manual': 4000,
}
```

```python
df = pd.DataFrame.from_dict(data, orient='index', columns=['Time'])
df
```

```python
ax = df.plot(kind='bar')
ax.set_xticklabels(data.keys(), rotation=360)
ax.get_legend().remove()
plt.ylabel('Time per 2.4M steps (seconds)')
plt.yticks([0, 1000, 2000, 3000, 4000])
plt.title('OpenMM 8.3: hybrid topology performance analysis')
plt.savefig('openmm8.3_hybrid_top_analysis.png', bbox_inches='tight')
```
